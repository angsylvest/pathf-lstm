```python
from main import *
from matplotlib import pyplot as plt
```

```python
def plotter(metric, title, auctioneer_type):
    plt.clf()
    
    plt.figure(figsize=(10, 5), dpi=300)

    plt.bar(range(len(metric)), metric)
    average_metric = sum(metric) / len(metric)
    plt.axhline(y=average_metric, color='r', linestyle='-', label=f'Average {title} = {average_metric:.2f}')

    plt.xlabel('Test number')
    plt.xticks(range(len(metric)), range(1, len(metric) + 1))
    plt.ylabel(title)
    plt.title(f'{title} for {auctioneer_type} auctioneer')
    
    plt.legend()
    plt.show()
```

```python
# flowtimes, makespans = run(auctioneer_type='random', num_agents=20, horizon=40, num_tests=30, env_shape=(20, 20))

plotter(makespans, 'Makespan', 'random')
plotter(flowtimes, 'Flowtime', 'random')
```

```python
# Make sure to rerun file to have the same random seed
flowtimes_max, makespans_max = run(auctioneer_type='max', num_agents=20, horizon=40, num_tests=30, env_shape=(20, 20))

plotter(makespans_max, 'Makespan', 'bid maximizer')
plotter(flowtimes_max, 'Flowtime', 'bid maximizer')
```

```python
flowtimes_min, makespans_min = run(auctioneer_type='min', num_agents=20, horizon=40, num_tests=30, env_shape=(20, 20))

plotter(makespans_min, 'Makespan', 'bid minimizer')
plotter(flowtimes_min, 'Flowtime', 'bid minimizer')
```
